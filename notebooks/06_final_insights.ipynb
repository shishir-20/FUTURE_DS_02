# Final Business Insights

## Insight 1: Month-to-Month Contracts Show Highest Churn

Observation:
Customers using month-to-month contracts leave more frequently than customers with yearly contracts.

Recommendation:
Offer discounts, rewards, or incentives for long-term subscriptions.

---

## Insight 2: New Customers Leave Earlier

Observation:
Customers with shorter tenure show higher churn rates.

Recommendation:
Improve onboarding experience and engage customers early.

---

## Insight 3: High Monthly Charges Increase Churn

Observation:
Customers paying higher monthly fees appear more likely to leave.

Recommendation:
Create value bundles and pricing offers.

---

## Insight 4: Electronic Check Users Have High Churn

Observation:
Customers using electronic check payment methods show higher churn.

Recommendation:
Review payment process and identify friction points.

---

## Insight 5: Long-Term Customers Have Better Retention

Observation:
Retention improves significantly as customer tenure increases.

Recommendation:
Introduce loyalty rewards and customer retention programs.
